```python
#Import Libraries
import pandas as pd
```

```python
#Open Dict File
file = open('summary.txt','r').read()
user_dict = eval(file)
```

```python
data = pd.DataFrame.from_dict({(i,j,k): user_dict[i][j][k]
                           for i in user_dict.keys() 
                           for j in user_dict[i].keys()
                           for k in user_dict[i][j].keys()},
                           #for l in user_dict[i][j][k]},
                       orient='index')
print(data)
```
