```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import descartes
from shapely.geometry import Point, Polygon
import geopandas as gpd
import pysal as ps

%matplotlib inline
```

```python
# Create the map
npu_map = gpd.read_file('NPU.shp')
fig,ax = plt.subplots(figsize = (15, 15))
npu_map.plot(ax = ax)
```

```python
# Call up the csv dataset
df = pd.read_csv('cobra-clean2.csv')
crs = {'init': 'epsg:4326'}
df.head()
```

```python
# Creates a list of points out longitude and latitude. We can drop the Longitude and Latitude columns now
geometry = [Point(xy) for xy in zip(df["Longitude"], df["Latitude"])]
geometry[:3]
```

```python
geo_df = gpd.GeoDataFrame(df, # Specify our data
                         crs = crs, # Specify our coordinate reference system
                         geometry = geometry) # Specify the geometry list we created
geo_df.head()
```

```python
# We now layer our points on top of the map
fig,ax = plt.subplots(figsize = (15, 15))
npu_map.plot(ax = ax, alpha=0.4, color="grey")
# Going to label using Shift Occurrence
geo_df[geo_df['Shift Occurence'] == 'Day'].plot(ax=ax, markersize=4, color="blue", marker="o", label="Day")
geo_df[geo_df['Shift Occurence'] == 'Evening'].plot(ax=ax, markersize=4, color="black", marker="^", label="Evening")
geo_df[geo_df['Shift Occurence'] == 'Morning'].plot(ax=ax, markersize=4, color="yellow", marker="*", label="Morning")
plt.legend(prop={'size': 15})
```
